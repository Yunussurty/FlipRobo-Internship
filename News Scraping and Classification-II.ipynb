## News Scraping and Classification II - Sep2019

#### We will be scraping the news from India times for the month of September 2019

```python
#importing the libaries. We will be using selenium webdriver for this

import pandas as pd
import selenium
from selenium import webdriver
import time
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.desired_capabilities import DesiredCapabilities
```

```python
#creating the input function for the data to be scrape

a=input('Enter the year you want to scrap the new for: ')
b=input('Enter the month in digit only ,i.e for Jan enter 1 or for Nov enter 11: ')
c=input('The number of days the month have: ')
url='https://timesofindia.indiatimes.com/archive/year-{},month-{}.cms'.format(a,b)
```

```python
print(url)
```

```python
driver=webdriver.Chrome('chromedriver.exe')
driver.get(url)
driver.maximize_window()
time.sleep(2)

#1st we will scrape all the dates url
date_urls=[]
dates=[]

for i in driver.find_elements_by_xpath("//td[@align='center']/a"):
    date_urls.append(i.get_attribute('href'))
for i in date_urls:
    if i not in dates:
        dates.append(i)
print(len(dates))
```

```python
if len(dates)==int(c):
    print('All ok')
else:
    print('Recheck code')
```

```python
# creating the empty list for date to be store
news_date=[]
news_author=[]
news_topic=[]
news_headlines=[]
news_description=[]
```

```python
#creating the function

capa = DesiredCapabilities.CHROME
capa["pageLoadStrategy"] = "eager" #since economictimes has long loading time. This command will start scraping once 
driver = webdriver.Chrome(desired_capabilities=capa)  #the html is loaded.

new_dates=[]
for m in range(1,len(dates)+1):
    driver.get(dates[m-1])
    
    for j in driver.find_elements_by_xpath('//span[@style="font-family:arial ;font-size:12;color: #006699"]/a'):
        new_dates.append(j.get_attribute('href')) #appending all the news link of the particular day
    for k in new_dates:
        driver.get(k)
        
        try:
            try:news_date.append(driver.find_element_by_xpath('//div[@data-plugin="dateformat"]').text.replace("Created: ",'').replace("Updated: ",''))
            except:news_date.append(driver.find_element_by_xpath('//div[@class="yYIu- byline"]/span').text.replace("Created:",'').replace("Updated: ",''))
        except NoSuchElementException:
            news_date.append("--")        
        
        try:
            try:news_author.append(driver.find_element_by_xpath('//div[@class="yYIu- byline"]').text.split("/")[0])
            except:news_author.append(driver.find_element_by_xpath('//div[@class="as_byline"]/div').text)
        except NoSuchElementException:
            news_author.append("--")
            
        try:
            try:news_topic.append(driver.find_element_by_xpath('//ul[@itemtype="https://schema.org/BreadcrumbList"]/li[2]/a/span').text)
            except:
                try:news_topic.append(driver.find_element_by_xpath('//ul[@itemtype="https://schema.org/BreadcrumbList"]/li[1]/a/span').text)
                except:news_topic.append(driver.find_element_by_xpath('//ol[@itemtype="https://schema.org/BreadcrumbList"]/li[2]/a/span').text)
        except NoSuchElementException:
            news_topic.append("--")    
            
        try:
            try:news_headlines.append(driver.find_element_by_xpath('//h1[@class="_1Y-96"]/span').text)
            except:news_headlines.append(driver.find_element_by_xpath('//div[@class="as_heading"]/h1/arttitle').text)
        except NoSuchElementException:
            news_headlines.append("--")
            
        try:
            try:news_description.append(driver.find_element_by_xpath('//div[@class="_3YYSt clearfix  "]').text.replace('\n',''))
            except:news_description.append(driver.find_element_by_xpath('//div[@class="Normal"]').text.replace('\n',''))
        except NoSuchElementException:
            news_description.append("--")
        print((len(news_date),len(news_author),len(news_topic),len(news_headlines),len(news_description)))
    
    print('Scraping of date', m, 'completed')
    new_dates=[]
    
```

```python
#lets create the data frame
df=pd.DataFrame({"Date and time of upload":news_date,"Author name":news_author,"Topic or Vertical":news_topic,
                 "Headlines":news_headlines,"Description":news_description})    
```

```python
#removing the null values from the dataframe
index_names = df[ df['Description'] == "--" ].index
df.drop(index_names, inplace = True)
```

```python
#creating the csv for scrape date
df.to_csv('NewsSep19.csv')
```

```python
#creating the JSON file
df.to_json('NewsSep19.json', orient = 'split', compression = 'infer', index = 'true')
```

```python
#creating the excel file
df.to_excel('NewsSep19.xlsx')
```
