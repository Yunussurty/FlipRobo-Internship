#### 1. Scrape the details of most viewed videos on YouTube from Wikipedia:
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos/
You need to find following details:
A) Rank
B) Name
C) Artist
D) Upload date
E) Views

```python
#loading the drivers and URL

driver=webdriver.Chrome('chromedriver.exe')
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
driver.maximize_window()
time.sleep(2)

#creating the empty list
Rank=[]
Name=[]
Artist=[]
Upload=[]
Views=[]

#scraping the Rank
for i in driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[1]"):
    Rank.append(i.text)

#scraping the Name
for i in driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[2]"):
    Name.append(i.text.split('[')[0])

#scraping the Artist
for i in driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[3]"):
    Artist.append(i.text)

#scraping the upload date
for i in driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[5]"):
    Upload.append(i.text)

#scraping the number of views
for i in driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[4]"):
    Views.append(i.text)
    
Yotube_Most_Views={'Rank':Rank,'Name':Name,'Artist':Artist,'Uploaded Date':Upload,'Views':Views}
Yotube_Most_Views=pd.DataFrame(data=Yotube_Most_Views)
Yotube_Most_Views
```

#### 2.Scrape the details team India’s international fixtures from bcci.tv.
Url = https://www.bcci.tv/.
You need to find following details:
A) Match title (I.e. 1st ODI)
B) Series
C) Place
D) Date
E) Time

```python
#loading the drivers and url
driver=webdriver.Chrome('chromedriver.exe')
driver.get('https://www.bcci.tv/')
driver.maximize_window()

#clickin on International tab
driver.find_element_by_xpath("//div[@class='navigation__drop-down drop-down drop-down--reveal-on-hover']").click()

#clicking on Fixtures
driver.find_element_by_xpath("//a[@class='navigation__link navigation__link--in-drop-down']").click()

#creating the empty list
Title=[]
Series=[]
Place=[]
Date=[]
Time=[]

#scraping the title
for i in driver.find_elements_by_xpath("//p[@class='fixture__additional-info']/strong"):
    Title.append(i.text.replace('\n',' '))

#scraping the series
for i in driver.find_elements_by_xpath("//div[@class='fixture__teams']"):
    Series.append(i.text.replace('\n',' '))

#scraping the place
for i in driver.find_elements_by_xpath("//p[@class='fixture__additional-info']/span"):
    Place.append(i.text.replace('\n',' '))
    
#scraping the date
date= driver.find_elements_by_xpath("//div[@class='fixture__full-date']")
for i in range(len(date)):
    Date.append(date[i].text.split('\n')[0] + date[i].text.split('\n')[1])

#scraping the time
for i in driver.find_elements_by_xpath("//span[@class='fixture__time']"):
    Time.append(i.text)
    
#creating the DataFrame
Fixtures={'Title':Title,'Series':Series,'Place':Place,'Date':Date,'Time':Time}
Fixtures=pd.DataFrame(data=Fixtures)
Fixtures
```

#### 4. Scrape the details of State-wise GDP of India from statisticstime.com.
Url = http://statisticstimes.com/
You have to find following details:
A) Rank
B) State
C) GSDP at current price (19-20)
D) GSDP at current price (18-19)
E) Share(18-19)
F) GDP($ billion)
Note: - From statisticstimes home page you have to reach to economy page through code.

```python
#loading the drivers and url
driver=webdriver.Chrome('chromedriver.exe')
driver.get('http://statisticstimes.com/')
driver.maximize_window()

#getting to the Indian page under economy tab
page=driver.find_element_by_xpath("//div[@class='dropdown-content']/a[3]")
driver.get(page.get_attribute('href'))

#clicking on GDP of indian states
driver.find_element_by_xpath("//ul[@style='list-style-type:none;margin-left:20px;']/li[1]/a").click()

#creating the empty list
Rank=[]
state=[]
GSDP1=[]
GSDP2=[]
share=[]
GDP=[]

#scraping for name
for i in driver.find_elements_by_xpath("//*[@id='table_id']/tbody/tr/td[1]"):
    Rank.append(i.text)

#scraping for state name
for i in driver.find_elements_by_xpath("//*[@id='table_id']/tbody/tr/td[2]"):
    state.append(i.text)

#scraping for GSDP1
for i in driver.find_elements_by_xpath("//*[@id='table_id']/tbody/tr/td[3]"):
    GSDP1.append(i.text)

#scraping for GSDP1
for i in driver.find_elements_by_xpath("//*[@id='table_id']/tbody/tr/td[4]"):
    GSDP2.append(i.text)
    
#scraping for share
for i in driver.find_elements_by_xpath("//*[@id='table_id']/tbody/tr/td[5]"):
    share.append(i.text)
    
#scraping for GDP
for i in driver.find_elements_by_xpath("//*[@id='table_id']/tbody/tr/td[6]"):
    GDP.append(i.text)
```

```python
#creating the DataFrame
GDP_State={'Rank':Rank,'State':state,'GSDP 19-20':GSDP1,'GSDP 18-19':GSDP2,'Share 18-19':share,'GDP':GDP}
GDP_State=pd.DataFrame(data=GDP_State)
GDP_State
```

#### 5. Scrape the details of trending repositories on Github.com.
Url = https://github.com/
You have to find the following details:
A) Repository title
B) Repository description
C) Contributors count
D) Language used

```python
#loading the drivers and url
driver=webdriver.Chrome('chromedriver.exe')
driver.get('https://github.com/')
driver.maximize_window()

#clicking on trending
trending=driver.find_element_by_xpath('/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/div/ul[2]/li[3]/a')
driver.get(trending.get_attribute('href'))

#creating the empty list
name=[]
url=[]
desc=[]
count=[]
lang=[]

#getting the url of all the repository
for i in driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']/a"):
    url.append(i.get_attribute('href'))

#scraping the name
for i in driver.find_elements_by_xpath("//span[@class='text-normal']"):
    name.append(i.text.replace(' /',''))
                
for i in driver.find_elements_by_xpath("//p[@class='col-9 color-text-secondary my-1 pr-4']"):
    desc.append(i.text)
 
 #looping through urls to scrape other details

for i in url:
    driver.get(i)

#scraping for count
    count1=driver.find_elements_by_xpath("//div[@class='BorderGrid-cell']/h2/a/span")
    try:
        count.append((count1[-1].text))
    except:
        count.append('-')
        
#scraping for language
    language= driver.find_elements_by_xpath("//ul[@class='list-style-none']")
    try:
        lang.append((language[-1].text.replace('\n',' ')))
    except:
        lang.append('-')

#creating the dataframe
Github={'Repository Title':name,'Repository description':desc,'Contributors count':count,'Language used':lang}
Github=pd.DataFrame(data=Github)
Github
```

#### 7. Scrape the details of Data science recruiters from naukri.com.
Url = https://www.naukri.com/
You have to find the following details:
A) Name
B) Designation
C) Company
D) Skills they hire for
E) Location
Note: - From naukri.com homepage click on the recruiters option and the on the search pane type Data science and click on search. All this should be done through code

```python
#loading the drivers and url
driver=webdriver.Chrome('chromedriver.exe')
driver.get('https://www.naukri.com/')
driver.maximize_window()

#clicking on the recruiters page
Recruiters=driver.find_element_by_xpath('//*[@id="root"]/div[1]/div/ul[1]/li[2]/a')
driver.get(Recruiters.get_attribute('href'))

#entering the Data science in search bar and clicking on search
driver.find_element_by_xpath("//input[@class='sugInp']").send_keys('Data Science')
driver.find_element_by_xpath("//button[@class='fl qsbSrch blueBtn']").click()

#few recuiters have not given the location, hence filtering out those with the selection of location name.
for i in range(1,6):
    driver.find_element_by_xpath('//*[@id="acord"]/div[1]/a[{}]'.format(i)).click() #select all the location on front page

driver.find_element_by_xpath("//*[@id='acord']/div[1]/span").click() #click on more under location

#selecting all the remaining location
for i in range(1,22):
    driver.find_element_by_xpath('//*[@id="clstr_Recruiters Location"]/div/div[1]/div/a[{}]'.format(i)).click()

driver.find_element_by_xpath("//button[@class='blueBtn']").click() # clicking on refine

#now we have all the location details.

#creating the empty list
name=[]
designation=[]
company=[]
skills=[]
location=[]

#creating the range function to scrap from all the pages
for i in range(0,117):
    for i in driver.find_elements_by_xpath("//span[@class='fl ellipsis']"): #scrapping for name
        name.append(i.text)

    for i in driver.find_elements_by_xpath("//span[@class='ellipsis clr']"): #scrapping for designation
        designation.append(i.text)

    for i in driver.find_elements_by_xpath("//p[@class='highlightable']/a[2]"): #scrapping for company name
        company.append(i.text)

    for i in driver.find_elements_by_xpath("//div[@class='hireSec highlightable']"): #scrapping for skills they hire
        skills.append(i.text)

    for i in driver.find_elements_by_xpath("//p[@class='highlightable']/span[2]/small"): #scrapping for location
        location.append(i.text)

#to click on next button
    try:
        try:
            driver.find_element_by_xpath("//div[@class='pagination']/button[2]").click()
        except:
            driver.find_element_by_xpath("//div[@class='pagination']/button").click()
    except:
        NoSuchElementException
```

```python
#creating the dataframe for 1000 recuiters
Recruiters={'Name':name[:1000],'Designation':designation[:1000],'Company':company[:1000],
            'Skill I hire for':skills[:1000],'Location':location[:1000]}
Recruiters=pd.DataFrame(data=Recruiters)
Recruiters
```

#### 10. Details of Datasets from UCI machine learning repositories.
Url = https://archive.ics.uci.edu/
You have to find the following details:
A) Dataset name
B) Data type
C) Task
D) Attribute type
E) No of instances
F) No of attribute
G) Year
Note: - from the home page you have to go to the Show All Dataset page through code.

```python
#loading the drivers and url
driver=webdriver.Chrome('chromedriver.exe')
driver.get('https://archive.ics.uci.edu/')
driver.maximize_window()
time.sleep(4)

#clicking on vew all dataset
dataset=driver.find_element_by_xpath("/html/body/table[1]/tbody/tr/td[2]/span[2]/a")
driver.get(dataset.get_attribute('href'))

#creating the empty list
name=[]
datatype=[]
task=[]
attribute=[]
instances=[]
no_attribute=[]
year=[]

#scraping all the details
for i in driver.find_elements_by_xpath("//tr['@bgcolor=#003366']/td/table/tbody/tr/td[2]/p/b/a"):
    name.append(i.text)

for i in driver.find_elements_by_xpath("//table[@border='1']/tbody/tr/td[2]/p"):
    datatype.append(i.text)
datatype=datatype[1:]

for i in driver.find_elements_by_xpath("//table[@border='1']/tbody/tr/td[3]/p"):
    task.append(i.text)
task=task[1:]

for i in driver.find_elements_by_xpath("//table[@border='1']/tbody/tr/td[4]/p"):
    attribute.append(i.text)
attribute=attribute[1:]

for i in driver.find_elements_by_xpath("//table[@border='1']/tbody/tr/td[5]/p"):
    instances.append(i.text)
instances=instances[1:]

for i in driver.find_elements_by_xpath("//table[@border='1']/tbody/tr/td[6]/p"):
    no_attribute.append(i.text)
no_attribute=no_attribute[1:]

for i in driver.find_elements_by_xpath("//table[@border='1']/tbody/tr/td[7]/p"):
    year.append(i.text)
year=year[1:]

#creating the dataset
UCI={'Dataset Name':name,'Data Type':datatype,'Task':task,'Attribute Type':attribute,
    'No of Instances':instances,'No of Attribue':no_attribute,'Year':year}
UCI=pd.DataFrame(data=UCI)
UCI
```
